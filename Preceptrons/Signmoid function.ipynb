```python
import numpy as np
import matplotlib.pyplot as plt 


# signmoid function is used to calculate the probability of next aviliable points.
```

```python
def drawLine(x1, x2):
    line = plt.plot(x1,x2)
    
def sigmoid(score):
    return 1/(1+np.exp(-score))
```

```python
n_pts = 100
np.random.seed(0)

#np.random.normal()   generates a list of number based on standard normal distribution
#normal('centre of the disturbution', "Standard deviation aka how spread the data", #to be generated)
top_random_x1_values = np.random.normal(10, 2, n_pts)  #x coordinates of points 
top_random_x2_values = np.random.normal(12, 2, n_pts)  #y coordinates of points

bias = np.ones(n_pts)  #create an array of 1's with size of n_pts --100

#read np.transpose(...) 
#combined 2 1-D arrays into 1 2-Darray  
#e.g combined_array[][0] is equal to random_x1_values
#e.g combined_array[][1] is equal to random_x2_values

top_region = np.array([top_random_x1_values, top_random_x2_values, bias]).T  


# print(random_x1_values)
# print(random_x2_values)
# print(top_region)

bottom_random_x1_values = np.random.normal(5, 2, n_pts)  #x coordinates of points 
bottom_random_x2_values = np.random.normal(5, 2, n_pts)  #y coordinates of points
bottom_region = np.array([bottom_random_x1_values, bottom_random_x2_values, bias]).T

all_points = np.vstack((top_region, bottom_region))  #store 2 arrays into one stack and top_region will be on top of  the stack

#print(all_points)

w1 = -0.2
w2 = -0.35    
b = 3.5

line_parameters = np.matrix([w1, w2, b]).T # will be used in the following matrix multiplication
print(line_parameters)


#x1[0] represents the x-cooridinate of leftmost point of bottom region 
# and x1[1] represents x-cor the rightmost point of top_region in the graph 
#x2[0] represents the y-cooridinate of leftmost point of bottom region
# and x2[1] represents y-cor the rightmost point of top_region in the graph

x1 = np.array([bottom_region[:,0].min(), top_region[:,0].max()])

# 0 = w1x1 + w2x2 + b
# x2 = (- b - w1x1) / w2
x2 =  (-b - w1 * x1) / w2
print(x1)
print(x2)
print(x1, x2)

# Matrix multiplication comes in place 
print(all_points.shape)
print(line_parameters.shape)

linear_combined = all_points * line_parameters


print(linear_combined)

prob =  sigmoid(linear_combined)

print(prob)


```

```python

_, axes = plt.subplots(figsize=(4,4)) #figure size, could be any number
axes.scatter(top_region[:,0], top_region[:,1], color='red')
axes.scatter(bottom_region[:,0], bottom_region[:,1], color='blue')
drawLine(x1, x2)
plt.show()
```
